```python
tele = pd.get_dummies(data=tele, columns=['state', 'international_plan', 'voice_mail_plan', 'churn'], drop_first=True)
```

```python
tele.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(tele.drop(['area_code', 'phone_number', 'churn_True'], axis=1), 
                                                    tele['churn_True'], test_size=0.3, random_state=9, 
                                                    stratify=tele['churn_True'])
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
```

```python
dt.fit(X_train, y_train)
```

```python
dt.score(X_train, y_train)
```

```python
dt_pred = dt.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
print(classification_report(y_test, dt_pred))
print('\n')
print(confusion_matrix(y_test, dt_pred))
print('\n')
print(accuracy_score(y_test, dt_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=500)
```

```python
rf.fit(X_train, y_train)
```

```python
rf.score(X_train, y_train)
```

```python
rf_pred = rf.predict(X_test)
```

```python
print(classification_report(y_test, rf_pred))
print('\n')
print(confusion_matrix(y_test, rf_pred))
print('\n')
print(accuracy_score(y_test, rf_pred))
```
